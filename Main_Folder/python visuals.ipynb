```python
import matplotlib.pyplot as plt
import seaborn as sns

top_10_richest = df.sort_values(by="Net Worth", ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Net Worth", y='Name', data=top_10_richest, palette='viridis')
plt.xlabel('Net Worth (Billions $)')
plt.ylabel('Name')
plt.title('Top 10 Richest People in the World')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Billionaires')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Industry', y='Net Worth', data=df, palette='Set2')
plt.xlabel('Industry')
plt.ylabel('Net Worth (Billions $)')
plt.title('Net Worth Distribution by Industry')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Net Worth', data=df, hue='Country', palette='viridis')
plt.xlabel('Age')
plt.ylabel('Net Worth (Billions $)')
plt.title('Age vs. Net Worth of Billionaires')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
source_net_worth = df.groupby('Source')[['Net Worth']].sum().sort_values(by='Net Worth', ascending=False).head(10)

# Create a horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(source_net_worth.index, source_net_worth['Net Worth'], color='gray')

ax.set_xlabel('Total Net Worth ($ Billion)')
ax.set_ylabel('Source')
plt.title('Top 10 Sources with the Highest Total Net Worth')

plt.gca().invert_yaxis()

plt.show()
```

```python
country_counts = df['Country'].value_counts().head(10)

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)

circle = plt.Circle((0, 0), 0.70, fc='Teal')
fig.gca().add_artist(circle)

ax.axis('equal')
plt.title('Top 10 Countries with the Most Rich Individuals', loc='center', fontsize=16)
plt.show()
```
